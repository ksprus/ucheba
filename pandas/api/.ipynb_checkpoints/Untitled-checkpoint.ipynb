## Задание 1

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

KEYWORDS = ['python', 'парсинг']
page = 1
fram =[]

def pages(self,fram):
    for i in KEYWORDS:
        word=i
        url='https://habr.com/ru/search/page'+str(page)+'/?q='+word+'&target_type=posts&order=relevance'
        req=requests.get(url)
        soup = BeautifulSoup(req.text)
        posts = soup.find_all('article', 'tm-articles-list__item')
        for post in posts:        
            a=post.text.lower()
            if word in a:
                fr= [{'Дата': post.time.text, 'Заголовок': post.h2.span.text,'Ссылка':'https://habr.com'+(post.h2.a.get('href'))}]
                fram+=fr
    return fram

while page < 5: #Здесь пишем сколько страниц с результатами хотим охватить
    frame=pages(page,fram)
    page +=1

table=pd.DataFrame(list(fram))
table.head(len(table))
```
