## Задание 1

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

req = requests.get('https://habr.com/ru/all')
soup = BeautifulSoup(req.text)
KEYWORDS = ['python', 'парсинг']

def table():
    data=pd.DataFrame()
    posts = soup.find_all('article', class_='post')
    for post in posts:
        post_id = post.parent.attrs.get('id')
        if not post_id:
            continue
        post_id = int(post_id.split('_')[-1])
        hubs = post.find_all('a', class_='hub-link')
        for hub in hubs:
            hub_lower = hub.text.lower()
            if any([hub_lower in desired for desired in KEYWORDS]):
                date = post.find('span', class_='post__time')
                title_element = post.find('a', class_='post__title_link')
                data=data.append({'Date':date.text, 'Name':title_element.text, 'Link':title_element.attrs.get('href') }, ignore_index=True)
    return data
data=table()
data=data.loc[:,['Date','Name','Link']]
data.head()
```
