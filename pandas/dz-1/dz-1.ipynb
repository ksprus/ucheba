## Задание 2

```python
import pandas as pd
data = pd.read_csv('power.csv')
filter_countries=data[(data['country']=='Latvia') | (data['country']=='Lithuania') | (data['country']=='Estonia')]
filter_countries_year=filter_countries.sort_values('year')
filter_countries_years=filter_countries_year[filter_countries_year['year'].isin(['2005', '2006', '2007', '2008', '2009', '2010'])]
def category(numb):
    filter_category = filter_countries_years[filter_countries_years['category'].isin([(numb)])]
    res=0
    for i in filter_category.quantity:
        if i > 0:
            res+=i
    print('Для категории',numb,'потребление стран -',int(res))
category(4) 
category(12) 
category(21)


```
