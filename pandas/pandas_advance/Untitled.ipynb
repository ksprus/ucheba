```python
import pandas as pd

def view():
    rows=[]
    with open('visit_log.csv','r') as f:
        a=f.readline().split(';')
        for i in f:
            b=i.split(';')
            b=[line.rstrip() for line in b]
            rows+=[b]
    columns=[line.rstrip() for line in a]
    logdata=pd.DataFrame(rows,columns=columns)
    f.close()
    logdata['source_type']=logdata['traffic_source']
    return logdata

logdata=view()

def source(logdata):
    def res():
        result=[]
        for i in logdata:
            a=[]
            result=[]
            for i in logdata:
                a+=[i]
                if len(a)>5:
                    result+=[a]
                    a=[]
        return result
    result= res()
    for a in result:
        if 'yandex' in a or 'google' in a:
            return ('organic')
        elif a[3]=='Russia' and a[5]=='paid' in a:
            return ('ad')
        elif a[3]=='Russia' and a[5]=='email' in a:
            return ('ad')
        elif a[5]=='email' and 'Russia' not in a:
            return ('other')
        elif a[5]=='paid' and 'Russia' not in a:
            return ('other')
    else:
        return a[5]
        
logdata['source_type']=logdata.apply(source,axis=1)
logdata.tail(10)

```
