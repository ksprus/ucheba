## Задание 1

```python
import pandas as pd
films = pd.read_csv('movies.csv')
ratings=pd.read_csv('ratings.csv')
films['rating'] = ratings['rating']

def film_class(films):
    if films['rating'] <=2:
        return ('Низкий рейтинг')
    elif films['rating'] <=4 and films['rating']>2:
        return ('Средний рейтинг')
    elif films['rating'] > 4.5:
        return ('Высокий рейтинг')
    
films['class']=films.apply(film_class, axis=1)
films.head(10)
```

```python
import pandas as pd
strok = pd.read_csv('keywords.csv')
strok.head()

def region(strok):
    for i in strok['keyword'].split(' '):
        if 'тул' in i and 'тульч' not in i:
            return ('Центр')
        elif 'москв' in i:
            return ('Центр')
        elif 'ярославл' in i:
            return ('Центр')
        elif 'петербург' in i:
            return ('Северо-Запад')
        elif 'псков' in i:
            return ('Северо-Запад')
        elif 'мурманск' in i:
            return ('Северо-Запад')
        elif 'владивосток' in i:
            return ('Дальний Восток')
        elif 'сахалин' in i:
            return ('Дальний Восток')
        elif 'хабаровск' in i:
            return ('Дальний Восток')
    else:
        return ('Undefined')

strok['region']=strok.apply(region, axis=1)

#Проверка результата
strok.loc[98080:98085]
```
