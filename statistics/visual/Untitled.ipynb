```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data=pd.read_csv('country_wise_latest.csv')
datas=data.loc[7:12,['Country/Region','Confirmed',  'Deaths', 'Recovered']]
datas=datas[['Country/Region','Confirmed',  'Deaths', 'Recovered']].groupby('Country/Region').sum()
datas['Recovered'].plot(kind='pie',title='Статистика выздоровевших по странам',ylabel='',autopct='%1.0f%%', colors = ['#14FA52', '#6283FA', '#12DEAF','#61E7FA', '#CA84FA', '#E0CB2B' ] )
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data=pd.read_csv('country_wise_latest.csv')
datas=data.loc[:10,['Country/Region','Confirmed',  'Deaths', 'Recovered']]
datas=datas[['Country/Region','Confirmed',  'Deaths', 'Recovered']].groupby('Country/Region').sum()
datas['Deaths'].plot(kind='area',title='Смертность',xlabel='',ylabel='Кол-во человек', color='#484848')
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data=pd.read_csv('country_wise_latest.csv')
datas=data.iloc[40:49,[0,6,7]]
datas=datas[['Country/Region','New deaths',  'New recovered']].groupby('Country/Region').sum()
datas=datas.rename(columns={'New deaths':'Смерти',  'New recovered':'Выздоровления'})
datas.plot(kind='bar',title='Соотношение новых случаев смерти и выздоровления',xlabel='',ylabel='Кол-во человек', color={"Смерти": "red", "Выздоровления": "#31E752"})
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data=pd.read_csv('country_wise_latest.csv')
datas=data.loc[40:44,['Country/Region','Confirmed', 'Confirmed last week']]
datas=datas[['Country/Region','Confirmed', 'Confirmed last week']].groupby('Country/Region').sum()
datas=datas.rename(columns={'Confirmed':'Случаи на этой неделе',  'Confirmed last week':'Случаи на прошлой неделе'})
datas.plot(title='Подтвержденные случаи заболевания', xlabel='', figsize=(8,5))
```
